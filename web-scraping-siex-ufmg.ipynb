## Função de tratamento de strings

```python
def trata_html(input):
    return " ".join(input.split()).replace('> <', '><')
```

## Parâmetros da requisição

```python
# Form Data
params = {
    'page': '0',
    'idLista': '',
    'inicializaLista': '',
    'dataInicio': '01/10/2019',
    'dataTermino': '11/10/2019',
    'tipo': '',
    'membro': '',
    'codMembro': '',
    'palavraChave': 'divulgacao cientifica',
    'status': '0',
    'registro': '',
    'unidade': '',
    'departamento': '',
    'areaTematica': '0',
    'areaTematicaAfim': '0',
    'linhaExtensao': '0',
    'grandeArea': '0',
    'tipoBolsa': '0',
    'Submit': 'Pesquisar',
}
```

## Obtendo o total de páginas com resultados da pesquisa

```python
response = requests.post(url, headers = headers, data = params)
html = response.text
html = trata_html(html)
soup = BeautifulSoup(html, 'html.parser')
tamanho = []
for item in soup.find('div', {'class': 'pagebanner'}).get_text().split():
    if(item.isnumeric()):
        tamanho.append(item)
tamanho
```

```python
import math

pages = math.ceil(int(tamanho[-1]) / int(tamanho[-2]))
pages
```

## Obtendo os resultados da pesquisa

```python
tabela = []

for i in range(pages):
    params.update({'d-16544-p': str(i + 1)})
    response = requests.post(url, headers = headers, data = params)
    html = response.text
    html = trata_html(html)
    soup = BeautifulSoup(html, 'html.parser')

    table = soup.find('table', {'class': 'tabelaResultado', 'id': 'tabelaPesquisa'}).tbody.findAll('tr')

    for item in table:
        linha = []
        for i in item:
            linha.append(i.get_text())
        tabela.append(linha)
    
resultado = pd.DataFrame(tabela)
resultado
```

## Adicionando ao DataFrame as demais palavras-chave

```python
for index, page in enumerate(resultado[0]):
    url = 'https://sistemas.ufmg.br/siex/' + page
    response = urlopen(url)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    resultado.iloc[index, -1] = soup.find('td', text = 'Palavras-chave:').findNext().getText().strip()

resultado
```
